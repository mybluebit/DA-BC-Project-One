```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Imported 4 sets of data
wine_file_red = "red_wine_0to400_anyrating.csv"
wine_file_sparkling = "sparkling_wine_0to400_anyrating.csv"
wine_file_white = "white_wine_0to400_anyrating (1).csv"
wine_file = "winemag-data-130k-v2.csv"
wine_file_df = pd.read_csv(wine_file)
wine_file_red_df = pd.read_csv(wine_file_red)
wine_file_sparkling_df = pd.read_csv(wine_file_sparkling)
wine_file_white_df = pd.read_csv(wine_file_white)
wine_file_df.head()
```

```python
# merged red&white&sparkling and dropped NA values and got rid of Unnamed 0
merged_df = wine_file_red_df.append(wine_file_white_df)
merged_df_1 = merged_df.append(wine_file_sparkling_df)
merged_df_1
```

```python
merged_df_2 = merged_df_1.drop(['Unnamed: 0'], axis=1)
merged_df_2
```

```python
merged_df_3 = merged_df_2.dropna()
merged_df_3.head()
```

```python
# removed unnamed column and replaced United States with US
merged_df_3["Country"]= merged_df_3["Country"].replace("United States", "US")
merged_df_3
```
