```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
#Imported 4 sets of data
wine_file_red = "red_wine_0to400_anyrating.csv"
wine_file_sparkling = "sparkling_wine_0to400_anyrating.csv"
wine_file_white = "white_wine_0to400_anyrating (1).csv"
wine_file = "winemag-data-130k-v2.csv"
wine_file_df = pd.read_csv(wine_file)
wine_file_red_df = pd.read_csv(wine_file_red)
wine_file_sparkling_df = pd.read_csv(wine_file_sparkling)
wine_file_white_df = pd.read_csv(wine_file_white)
wine_file_df.head()
```

```python
# merged red&white&sparkling and dropped NA values and got rid of Unnamed 0
merged_df = wine_file_red_df.append(wine_file_white_df)
merged_df_1 = merged_df.append(wine_file_sparkling_df)
merged_df_1
```

```python
merged_df_2 = merged_df_1.drop(['Unnamed: 0'], axis=1)
merged_df_2
```

```python
merged_df_3 = merged_df_2.dropna()
merged_df_3.head()
```

```python
# removed unnamed column and replaced United States with US
merged_df_3["Country"]= merged_df_3["Country"].replace("United States", "US")
merged_df_3
```

```python
#Drop columns that we are not going to use
merged_df_4 = merged_df_3.drop(columns=["Name", "NumberOfRatings", "Winery"])
merged_df_4
```

```python
#Remove " View Shop" prices
merged_df_5 = merged_df_4[merged_df_4.Price.str.contains("View shops") == False]
merged_df_5
```

```python
#remove"$"
merged_df_5['Price'] = merged_df_5['Price'].str.replace('$','')
merged_df_5
```

```python
merged_df_5["Price"].unique()
```

```python
merged_df_6 = merged_df_5[merged_df_5.Price.str.contains("ratings") == False]
```

```python
merged_df_6["Price"].unique()
```

```python
# Convert price to integer
# merged_df_6['Rating'] = pd.to_numeric(merged_df_6['Rating'])
# merged_df_6.astype({'Price': 'int32'}).dtypes
# pd.to_numeric(merged_df_6["Price"], downcast='float')
# merged_df_6['Rating'].astype(str).astype(float)
merged_df_6["Rating"] = merged_df_6["Rating"].map(float)
merged_df_6["Price"] = merged_df_6["Price"].map(float)
```

```python
merged_df_6.head()
```

```python
#Multiply rating by 20
merged_df_6["Rating"] = 20 * merged_df_6["Rating"]
```

```python
merged_df_6
```

```python
#group by country, average price and rating
merged_df_7 = merged_df_6.groupby(['Country']).mean()
```

```python
merged_df_7 = merged_df_7.rename(columns={ "Rating": "Consumer Rating", "Price" : "Vivino Prices"})
```

```python
#Drop NA
# filter columns to Country, rating and price
#Rename columns
#Convert data if needed
wine_file_df.dropna()
new_wine = wine_file_df.loc[:, ["country", "points", "price"]]
new_wine_1 = new_wine.dropna(how='any')
new_wine_2 = new_wine_1.rename(columns={"country": "Country", "points": "Professional Rating", "price" : "Wine Cellar Prices"})
```

```python
new_wine_3 = new_wine_2.groupby(["Country"]).mean()
new_wine_3
```

```python
merged_df = pd.merge(new_wine_3, merged_df_7, on="Country")
merged_df
```
